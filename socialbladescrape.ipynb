```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import pandas as pd
from selenium.webdriver.chrome.options import Options
import time
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
htmlList1 = []
TotalGradeList=[]
YoutuberNameList=[]
YoutuberDateList=[]
SubscriberRankList=[]
VideoViewRankList=[]
SubscribersLast30List=[]
VideoViewsLast30List=[]
for i in YoutubersDict.keys():
    for j in YoutubersDict[i]:
        path="/Users/dawsonmcmahon/Desktop/Web_Data_Analytics/Files_Directory/chromedriver"
        driver = webdriver.Chrome(path, options = chrome_options)
        driver.get('https://web.archive.org/web/'+str(j)+"/https://socialblade.com/youtube/user/"+i)
        html = driver.page_source
        htmlList1.append(html) # Get the html of the page
        driver.quit() # Close the browser
        soup = bs(html, 'html.parser')  
        YoutuberNameList.append(i) #append username
        YoutuberDateList.append(j) #append date
        TotalGrade=soup.find(id="afd-header-total-grade") #get and append total grade
        TotalGradeList.append(TotalGrade)
        SubscriberRank= soup.find(id="afd-header-subscriber-rank")
        SubscriberRankList.append(SubscriberRank)
        VideoViewRank= soup.find(id="afd-header-videoview-rank")
        VideoViewRankList.append(VideoViewRank)
        SubscribersLast30= soup.find(id="afd-header-subs-30d")
        SubscribersLast30List.append(SubscribersLast30)
        VideoViewsLast30= soup.find(id="afd-header-views-30d")
        VideoViewsLast30List.append(VideoViewsLast30)
print(TotalGradeList)


```
